```python
import os
import xml.etree.ElementTree as ET
from tqdm import tqdm
from pathlib import Path

def convert_label(names, lb_path, lb_path_new):
    def convert_box(size, box):
        dw, dh = 1. / size[0], 1. / size[1]
        x, y, w, h = box
        return x * dw, y * dh, w * dw, h * dh

    with open(lb_path) as in_file:
        tree = ET.parse(in_file)

    with open(lb_path_new, 'w') as out_file:
        root = tree.getroot()
        size = root.find('size')
        w = int(size.find('width').text)
        h = int(size.find('height').text)

        for obj in root.iter('object'):
            cls = obj.find('name').text
            # xmlbox = obj.find('bndbox')
            # bb = convert_box((w, h), [float(xmlbox.find(x).text) for x in ('xmin', 'ymin', 'xmax', 'ymax')])
            xmlbox = obj.find('bndbox')
            if xmlbox is None:
                print(f"No bounding box in {lb_path}, skipping object.")
                continue
            b = (
                float(xmlbox.find('xmin').text) if xmlbox.find('xmin') is not None else None,
                float(xmlbox.find('ymin').text) if xmlbox.find('ymin') is not None else None,
                float(xmlbox.find('xmax').text) if xmlbox.find('xmax') is not None else None,
                float(xmlbox.find('ymax').text) if xmlbox.find('ymax') is not None else None
            )
            if None in b:
                continue
            bb = convert_box((w, h), b)
            occ = int(obj.find('occlusion').text)
            cls_id = names.index(cls)  # class id
            out_file.write(" ".join([str(a) for a in (cls_id, *bb, occ)]) + '\n')

path = Path('/home/hyu3/Choi/aue8088-pa2/kaist-rgbt')  # dataset root dir
with open(str(path / 'val_sets.txt'), 'r') as fp:
    labels = [f.replace('images/', '').replace('lwir', 'labels-xml').replace('.jpg', '.xml').rstrip('\n') for f in fp.readlines()]


with open(path / 'kaist-rgbt.names') as fp:
    names = [f.rstrip('\n') for f in fp.readlines()]

for label in tqdm(labels, total=len(labels)):
    lb_file = label
    lb_file_new = label.replace('labels-xml', 'labels').replace('.xml', '.txt')

    os.makedirs(os.path.dirname(lb_file_new), exist_ok=True)

    convert_label(names, lb_file, lb_file_new)  # convert labels to YOLO format
```

```python
import os
import xml.etree.ElementTree as ET
from tqdm import tqdm
from pathlib import Path

def convert_box(size, box):
    dw = 1. / size[0]
    dh = 1. / size[1]
    x = (box[0] + box[2]) / 2.0
    y = (box[1] + box[3]) / 2.0
    w = box[2] - box[0]
    h = box[3] - box[1]
    return x * dw, y * dh, w * dw, h * dh

def convert_label(names, lb_path, lb_path_new):
    try:
        tree = ET.parse(lb_path)
    except ET.ParseError as e:
        print(f"Error parsing {lb_path}: {e}")
        return

    with open(lb_path_new, 'w') as out_file:
        root = tree.getroot()
        size = root.find('size')
        if size is None:
            print(f"No size information in {lb_path}, skipping file.")
            return
        
        w = size.find('width')
        h = size.find('height')
        if w is None or h is None:
            print(f"No width/height information in {lb_path}, skipping file.")
            return
        w = int(w.text)
        h = int(h.text)

        for obj in root.iter('object'):
            cls = obj.find('name').text if obj.find('name') is not None else 'unknown'
            if cls not in names:
                continue
            cls_id = names.index(cls)
            xmlbox = obj.find('bndbox')
            if xmlbox is not None:
                b = (float(xmlbox.find('xmin').text) if xmlbox.find('xmin') is not None else 0,
                     float(xmlbox.find('ymin').text) if xmlbox.find('ymin') is not None else 0,
                     float(xmlbox.find('xmax').text) if xmlbox.find('xmax') is not None else 0,
                     float(xmlbox.find('ymax').text) if xmlbox.find('ymax') is not None else 0)
                bb = convert_box((w, h), b)
                occ = int(obj.find('occlusion').text) if obj.find('occlusion') is not None else 0
                out_file.write(" ".join([str(a) for a in (cls_id, *bb, occ)]) + '\n')

path = Path('/home/hyu3/Choi/aue8088-pa2/kaist-rgbt')  # dataset root dir
with open(str(path / 'val_sets.txt'), 'r') as fp:
    labels = [f.replace('images/', '').replace('lwir', 'labels-xml').replace('.jpg', '.xml').rstrip('\n') for f in fp.readlines()]

with open(path / 'kaist-rgbt.names') as fp:
    names = [f.rstrip('\n') for f in fp.readlines()]

for label in tqdm(labels, total=len(labels)):
    lb_file = str(path / label)  # 전체 경로 생성
    lb_file_new = str(path / label.replace('labels-xml', 'labels_converted').replace('.xml', '.txt'))

    os.makedirs(os.path.dirname(lb_file_new), exist_ok=True)

    convert_label(names, lb_file, lb_file_new)  # convert labels to YOLO format
```

```python
import numpy as np
from sklearn.cluster import KMeans

def compute_anchors(dataset, n_clusters=9):
    all_boxes = []
    for img, targets in dataset:
        for target in targets:
            _, _, _, w, h = target  # assuming target format [class, x_center, y_center, width, height]
            all_boxes.append([w, h])
    
    kmeans = KMeans(n_clusters=n_clusters, random_state=0).fit(all_boxes)
    anchors = kmeans.cluster_centers_
    return anchors

# Example dataset
dataset = [
    (np.random.rand(640, 480, 3), [[0, 0.5, 0.5, 0.2, 0.4], [1, 0.4, 0.4, 0.3, 0.3]]),
    # Add more data here
]

optimal_anchors = compute_anchors(dataset)
print("Optimal anchors:\n", optimal_anchors)
```
